```python
from graphviz import Digraph

diagram = Digraph(
    comment="E-Commerce Data Architecture",
    format="png"
)

diagram.attr(rankdir="LR")  # Left to Right flow

# Nodes
diagram.node("User", "Customer\n(Web / Mobile App)")
diagram.node("API", "Backend API")
diagram.node("OLTP", "Transactional DB\n(MySQL / PostgreSQL)")
diagram.node("Queue", "Message Queue\n(Kafka / RabbitMQ)")
diagram.node("Lake", "Data Lake\n(AWS S3 / GCS)")
diagram.node("Warehouse", "Data Warehouse\n(Redshift / BigQuery)")
diagram.node("BI", "Analytics & BI\nDashboards")

# Edges
diagram.edge("User", "API")
diagram.edge("API", "OLTP")
diagram.edge("API", "Queue")
diagram.edge("Queue", "Lake")
diagram.edge("Lake", "Warehouse")
diagram.edge("Warehouse", "BI")

# Save and render
diagram.render("ecommerce_data_architecture", view=True)
```
